```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from time import sleep
from random import randint
```

```python
Quote =[]
Page_number= []
Author = []
Tags = []
sub_Tags=[]
def check_module():
    #request will request the response from the url
    page = requests.get("https://quotes.toscrape.com/page/1/")
    # Beautiful soup is use to parse the html page
    soup = BeautifulSoup(page.text, 'html.parser')
    # Navig_data will find all li tag which have class next so we can acess next page
    navig_data= soup.findAll('li',attrs={'class':'next'})
    i = 1
    flag = True
    while(flag==True):
   
        page = requests.get("https://quotes.toscrape.com/page/"+str(i)+"/")
        #print(page)
        soup = BeautifulSoup(page.text, 'html.parser')
        navig_data= soup.findAll('li',attrs={'class':'next'})
        #print("navig_data===========",i,navig_data)
        if(len(navig_data) == 0):
            flag = False
        Quote_data = soup.findAll('div', attrs = {'class': 'quote'})
        Webscrab2 = crawl_module(Quote_data,i)
        #print("Response Massage of page number",i,"==>",page)
        #print("Quote Data",Quote_data)
        i=i+1
    else:
        pass
    return Webscrab2


def crawl_module(Quote_data,i):
    for ext in Quote_data:
        Page_number.append(str(i))
        # this will crawl all the required details which we expect from our question
        quot = ext.find('span', class_ = "text").text
        Quote.append(quot)
        Ath = ext.find('small', class_ = "author").text
        Author.append(Ath)
        Tagss = ext.find_all('meta', class_ = "keywords")
        for item in Tagss:
            link = item['content']
        #Tag = ext.find('meta', class_ = "keywords").keywords
        #print(Tag.strip())
        
        Tags.append(link)
    Webscrab = pd.DataFrame({"Page":Page_number,"Quotes" : Quote, "Author":Author, "Tags": Tags})
    return Webscrab
```
